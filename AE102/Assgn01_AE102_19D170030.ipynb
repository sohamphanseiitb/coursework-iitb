```python
# Q1 (i)
zval, N, i, mean, var = [], 5, 0, 0.5, (1/12)
stddev = np.sqrt(var)
mu, sigma = 0, 1

def x():
    return np.random.uniform(low = 0.0, high = 1.0, size = None)

def num(n):
    a = []
    for i in range(n):
        a.append(x())
    return a

def calmean():
    return d(num(N)).mean

def calvar():
    return d([x(), x(), x(), x(), x()]).variance

def calstd():
    return np.sqrt(calvar())

for i in range(1000):
    z = np.sqrt(N)*(calmean()-mean)/stddev
    zval.append(z)

# Q1 (i) --
plt.figure(figsize=(20, 10))
count, bins, ignored = plt.hist(zval, 100, density=True)
plt.plot(bins, norm(loc = mu, scale=sigma).pdf(bins),linewidth=2, color='r')
plt.title("Histogram of 'z' values at N = " + str(N))
plt.show()
```

## Q2 (iii)

```python
def r1():
  return np.random.randint(1, 12, size=None)

def r2():
  return np.random.randint(1, 20, size=None)

def isprime(n):
  for i in range(2, n):
    if n%i==0:
      return False
  return True

def p(b):
  if b==1:
    return (1/12)
  elif b==2:
    return (1/20)

def sum_array(start, end, array):
  sum = 0
  for i in range(start, end + 1):
    sum = sum + array[i-1]
  return sum 

def stochastic():
  store, prob = [], []
  for each in range(1000):
    store.append(isprime(r1() + r2()))
  for i in range(1, len(store)+1):
    prob.append(sum_array(1, i, store)/i)
  return prob[-1]

answer = []
for each in range(1000):
  answer.append(stochastic())

plt.figure(3)
count, bins, ignored  = plt.hist(answer, 50, density = True);
plt.title('Stochastic Process $n_T$ = 1000 throws for N = 1000 times')
plt.xlabel('Probability Value')

# Computing the mean and standard deviation of the above computed histogram
from scipy.stats import describe as d
a, b = d(answer).mean, d(answer).variance
plt.plot(bins, norm(loc = a, scale = np.sqrt(b)).pdf(bins), linewidth=2, color='r')
```
